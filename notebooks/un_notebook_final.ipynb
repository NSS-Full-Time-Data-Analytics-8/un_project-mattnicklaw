```python
gdp_df.columns=["Country", "Year", 'GDP_Per_Capita']
gdp_df
```

```python
internet_df.columns = ['Country', 'Year', 'Internet_Use']
internet_df
```

```python
gdp_and_internet_use = pd.merge(gdp_df, internet_df, how = "inner", on = ["Country", "Year"])
gdp_and_internet_use
```

```python
gdp_and_internet_use = gdp_and_internet_use.loc[(gdp_and_internet_use['Year'] == 2004) | (gdp_and_internet_use['Year'] == 2009) | (gdp_and_internet_use['Year'] == 2014)]
gdp_and_internet_use
```

```python
zero_four = gdp_and_internet_use.loc[(gdp_and_internet_use['Year'] == 2004)]
zero_four
```

```python
zero_nine = gdp_and_internet_use.loc[(gdp_and_internet_use['Year'] == 2009)]
zero_nine
```

```python
fourteen = gdp_and_internet_use.loc[(gdp_and_internet_use['Year'] == 2014)]
fourteen
```

```python
fourteen.sort_values(by='Internet_Use', ascending = False, inplace=True)
fourteen
```

In 2014, Iceland had the highest Internet_Use at 98.16%

In 2014, Timor-Leste had the lowest Internet_Use at 1.14%

```python
zero_nine.sort_values(by='Internet_Use', ascending = False, inplace=True)
zero_nine
```

```python
zero_four.sort_values(by='Internet_Use', ascending = False, inplace=True)
zero_four
```

In 2004, Sweden had the highest Internet_Use at 83.9%

In 2004, Myanmar had the lowest Internet_Use at 0.02%

```python
fourteen.sort_values(by='GDP_Per_Capita', ascending = False, inplace=True)
fourteen
```
